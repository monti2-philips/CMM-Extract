```python
import pandas as pd
import os

pd.set_option("display.max_columns",100)
```

```python
file = r"GROUP8 DATA.csv"
file_out = os.path.splitext(file)[0]+'.xlsx'
df = pd.read_csv(file)
```

```python
df.columns
```

```python
df.head()
```

```python
df_ = df.copy()[['Characteristic','K1 Measured value','K80 Sample','K7 Nest','K4 Time/Date', 'K2101 Nominal value', 'K2110 Lower limit', 'K2111 Upper limit', 'K2142 Unit']]
```

```python
df_.head()
```

```python
df_ = df_.astype({'Characteristic':'str',
            'K1 Measured value':'float64',
            'K80 Sample':'str',
            'K7 Nest':'str',
            'K4 Time/Date':'datetime64[ns]',
            'K2101 Nominal value':'float64',
            'K2110 Lower limit':'float64',
            'K2111 Upper limit':'float64',
            'K2142 Unit':'str'})
```

```python
df_pivot = df_.pivot_table(index=['K80 Sample', 'K7 Nest', 'K4 Time/Date'],columns='Characteristic',values='K1 Measured value')
df_pivot = df_pivot.reset_index()
df_pivot = df_pivot.sort_values(by="K4 Time/Date",ascending=True)
df_pivot
```

```python
df_pivot = df_pivot.drop_duplicates(subset=['K80 Sample','K7 Nest'], keep='last')
df_pivot = df_pivot[['K80 Sample', 'K7 Nest', 'K4 Time/Date',
                    'Q1 - Top Profile^Max', 'Q1 - Top Profile^Min',
                    'Q2 - Top Parallelism',
                    'Q3 - Left Tab Profile^Max', 'Q3 - Left Tab Profile^Min', 'Q3 - Right Tab Profile^Max', 'Q3 - Right Tab Profile^Min',
                    'Q4 - Top Width_X',
                    'Q5 - Top Position', 'Q5 - Top Position.X',
                    'Long Axis Position', 'Long Axis Position.Y', 'Long Axis Width_Y']]
df_pivot
```

```python
df_pivot.to_excel(file_out)
```
